# Tutorial: SymPy y NetworkX

Este notebook explica cómo usar **SymPy** para trabajar con matrices y cómo usar **NetworkX** para dibujar grafos.

Contiene ejemplos sencillos de:
- Crear matrices
- Obtener entradas, renglones y columnas
- Multiplicar matrices
- Elevar matrices a potencias
- Multiplicar vectores por matrices
- Calcular valores y vectores propios
- Dibujar grafos con NetworkX


```python
import sympy as sp
import networkx as nx
import matplotlib.pyplot as plt
```

## Crear matrices con SymPy

```python
A = sp.Matrix([
    [1,2],
    [3,4]
])

A
```

## Obtener entradas, renglones y columnas

```python
print('Entrada (0,1):', A[0,1])

print('Renglón 0:')
print(A.row(0))

print('Columna 1:')
print(A.col(1))
```

## Multiplicación de matrices

```python
B = sp.Matrix([
    [2,0],
    [1,5]
])

A * B
```

## Potencias de matrices

```python
C = sp.Matrix([
    [1,1],
    [0,1]
])

print('C^2')
print(C**2)

print('C^3')
print(C**3)
```

## Multiplicar vectores por matrices

```python
v = sp.Matrix([
    [1],
    [2]
])

A * v
```

## Valores y vectores propios

```python
M = sp.Matrix([
    [4,1],
    [2,3]
])

print('Valores propios:')
print(M.eigenvals())

print('\nVectores propios:')
print(M.eigenvects())
```

## Dibujar un grafo dirigido con NetworkX

```python
G = nx.DiGraph()

G.add_edge('A','B')
G.add_edge('A','C')
G.add_edge('B','C')
G.add_edge('C','A')

pos = nx.spring_layout(G)

nx.draw_networkx(G, pos, with_labels=True, arrows=True)

plt.title('Ejemplo de grafo dirigido')
plt.axis('off')
plt.show()
```
